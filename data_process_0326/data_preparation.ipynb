```python
import os
import pandas as pd
from tqdm.notebook import tqdm
from datetime import datetime
```

# 读取标签

```python
label = pd.read_feather("/home/datamake117/test0326/label_divide1_5.fea")
label['index'] = pd.to_datetime(label['index'])
label = label.set_index('index', drop=True)
label  # 1508 rows × 5319 columns (index: 2019-01-02 - 2025-03-04)
```

```python
label_ = pd.read_feather("/home/datamake117/test0326/label.fea")  # 使用不调整的label构造相关性矩阵
label_['index'] = pd.to_datetime(label_['index'])
label_ = label_.set_index('index', drop=True)
label_  # 1508 rows × 5319 columns (index: 2019-01-02 - 2025-02-21)
```

# 读取特征
## 读取特征0（1250个）

```python
year = 2020
os.chdir(r"/home/datamake117/test0326/feature0")
date_list = sorted(filter(lambda x: x[0] == "2" and int(x[:4]) >= year, os.listdir()))

factor_list = []
for file in tqdm(date_list):  # 将所有日期的因子数据合并
    dt = datetime.strptime(file[:-4], "%Y%m%d").date()
    factor = pd.read_feather(file)
    factor.fillna(0.0, inplace=True)
    factor.insert(loc=0, column="date", value=dt)
    factor_list.append(factor)

factor1 = pd.concat(factor_list)
factor1  # 5398930 rows × 1252 columns (index: 2020-01-02 - 2024-12-31)
```

## 读取特征1（1799个）

```python
year = 2020
os.chdir(r"/home/datamake117/test0326/feature1")
date_list = sorted(filter(lambda x: x[0] == "2" and int(x[:4]) >= year, os.listdir()))

factor_list = []
for file in tqdm(date_list):  # 将所有日期的因子数据合并
    dt = datetime.strptime(file[:-4], "%Y%m%d").date()
    factor = pd.read_feather(file)
    factor.fillna(0.0, inplace=True)
    factor.insert(loc=0, column="date", value=dt)
    factor_list.append(factor)

factor2 = pd.concat(factor_list)
factor2.columns = ['date', 'Code'] + [str(i) for i in range((len(factor1.columns) - 2), ((len(factor1.columns) - 2) + (len(factor2.columns) - 2)))]
factor2  # 5398930 rows × 1542 columns (index: 2020-01-02 - 2024-12-31)
```

## 合并全部特征（3049个）

```python
total_factor = pd.merge(factor1, factor2, how='inner', on=['date', 'Code'])
total_factor  # 5398930 rows × 2792 columns (index: 2020-01-02 - 2024-12-31)
```

# 合并特征和标签

```python
def look(date, Code):
    try:
        return label.loc[str(date), Code]
    except:
        return None
    
def look_(date, Code):
    try:
        return label_.loc[str(date), Code]
    except:
        return None

total_factor['label'] = [look(str(date), Code) for date, Code in zip(total_factor['date'], total_factor['Code'])]
total_factor['label_'] = [look_(str(date), Code) for date, Code in zip(total_factor['date'], total_factor['Code'])]
total_factor['date'] = total_factor['date'].astype("datetime64[ns]")
total_factor  # 5398930 rows × 2794 columns (index: 2020-01-02 - 2024-12-31)
```

# 增加可成交量变量

```python
total_factor['qcut'] = 0
total_factor['weight'] = 0
total_factor['qid'] = 0

daily_amount = pd.read_feather(r"/home/datamake117/test0326/can_trade_amt_1000.fea")
daily_amount['index'] = pd.to_datetime(daily_amount['index'], format='%Y%m%d')
daily_amount = daily_amount.set_index(['index'], drop=True)
daily_amount.index = daily_amount.index.astype('datetime64[ns]')
daily_amount.columns = daily_amount.columns.astype('str')
lookup_dict = daily_amount.stack().to_dict()
total_factor['amount'] = total_factor.set_index(['date', 'Code']).index.map(lookup_dict)
total_factor.to_pickle('/home/datamake117/data/haris/dataset_0326/total_factor.pkl')  # 因子+标签
total_factor  # 5398930 rows × 2798 columns (index: 2020-01-02 - 2024-12-31)
```

```python
# 设置两级索引，第一级为日期，第二级为股票代码
total_factor.set_index(["date", "Code"], inplace=True)
total_factor.sort_index(level=["date", "Code"], inplace=True)

# 重建 DataFrame，使每个日期的股票数相同，缺失值用 NaN 补充
all_dates = total_factor.index.get_level_values("date").unique()
all_stocks = total_factor.index.get_level_values("Code").unique()
full_index = pd.MultiIndex.from_product([all_dates, all_stocks], names=["date", "Code"])
total_factor = total_factor.reindex(full_index)
total_factor.sort_index(level=["date", "Code"], inplace=True)
total_factor.columns = [str(i) for i in range(total_factor.shape[1] - 6)] + ['label', 'label_', 'qcut', 'weight', 'qid', 'amount']
total_factor.index = total_factor.index.set_levels(total_factor.index.levels[0].strftime("%Y%m%d").astype(int), level='date')
total_factor  # 6354516 rows × 2796 columns (index: 2020-01-02 - 2024-12-31)
```
